# Use our train model to predict digits value from images

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2
import os
```

```python
model = tf.keras.models.load_model('./model/handwritten-tf.model.keras')
```

```python
image_index = 1
while os.path.isfile(f"digits/digit{image_index}.png"):
    try:
        img = cv2.imread(f"digits/digit{image_index}.png")[:,:,0]
        img = np.invert(np.array([img]))
        prediction = model.predict(img)
        print(prediction)
        print(f"This digit is probalbly a {np.argmax(prediction)}")
        plt.imshow(img[0], cmap=plt.cm.binary)
        plt.show()
    except e:
        print(f"An exception occurred for {image_index} image")
        print(e)
    finally:
        image_index += 1
```
